# Using Dask with TimeCopilot

```python
import nest_asyncio

nest_asyncio.apply()

from timecopilot import TimeCopilotForecaster

import dask.dataframe as dd
import pandas as pd

from timecopilot.models import SeasonalNaive
from timecopilot.models.foundation.chronos import Chronos
```

## Create the dataframe

```python
pd_df = pd.read_csv("https://timecopilot.s3.amazonaws.com/public/data/events_pageviews.csv", parse_dates=['ds'])
df = dd.from_pandas(pd_df)
df.head()
```

## Create a TimeCopilotForecaster

```python
tcf = TimeCopilotForecaster(
    models=[
        SeasonalNaive(),
        Chronos("autogluon/chronos-2-small")
    ]
)
```

## Create a Forecast

```python
result = tcf.forecast(
    df=df,
    h=12
)
```

```python
result_pd = result.compute()
```
