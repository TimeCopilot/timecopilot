# Using Dask DataFrames

TimeCopilot converts Dask DataFrames to pandas at entry so you can use them directly.


```python
import pandas as pd
import dask.dataframe as dd

from timecopilot import TimeCopilotForecaster
from timecopilot.models.stats import SeasonalNaive
```

```python
df = pd.read_csv(
    "https://timecopilot.s3.amazonaws.com/public/data/air_passengers.csv",
    parse_dates=["ds"],
)
dask_df = dd.from_pandas(df, npartitions=1)
```

```python
tcf = TimeCopilotForecaster(models=[SeasonalNaive()])
fcst_df = tcf.forecast(df=dask_df, h=12, freq="MS")
fcst_df.head()
```
