# Using Spark DataFrames

TimeCopilot converts Spark DataFrames to pandas at entry so you can use them directly.


```python
import pandas as pd
from pyspark.sql import SparkSession

from timecopilot import TimeCopilotForecaster
from timecopilot.models.stats import SeasonalNaive
```

```python
spark = SparkSession.builder.master("local[1]").appName("timecopilot").getOrCreate()
```

```python
df = pd.read_csv(
    "https://timecopilot.s3.amazonaws.com/public/data/air_passengers.csv",
    parse_dates=["ds"],
)
spark_df = spark.createDataFrame(df)
```

```python
tcf = TimeCopilotForecaster(models=[SeasonalNaive()])
fcst_df = tcf.forecast(df=spark_df, h=12, freq="MS")
fcst_df.head()
```

```python
spark.stop()
```
