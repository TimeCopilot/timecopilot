# Baseline Pandas Example

This is a reference point for using a TimeCopilotForecaster with pandas rather than with the alternative distributed dataframes like pyspark, ray, and dask.

```python
import nest_asyncio

nest_asyncio.apply()

from timecopilot import TimeCopilotForecaster

from timecopilot.models import SeasonalNaive
from timecopilot.models.foundation.chronos import Chronos

import pandas as pd
```

## Create the dataframe

```python
df = pd.read_csv("https://timecopilot.s3.amazonaws.com/public/data/air_passengers.csv", parse_dates=['ds'])
display(df.head())
```

## Create the TimeCopilotForecaster

```python
tcf = TimeCopilotForecaster(
    models=[
        SeasonalNaive(),
        Chronos()
    ]
)
```

## Generate a forecast

```python
result = tcf.forecast(
    df=df,
    h=12
)
```
